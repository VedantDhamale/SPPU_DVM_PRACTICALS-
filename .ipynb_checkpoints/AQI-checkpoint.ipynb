```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('AQI.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.dtypes
```

```python
df.describe()
```

```python
categorical_cols = ['stn_code','state', 'location', 'agency', 'type', 'location_monitoring_station']
for col in categorical_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
numerical_cols = ['so2', 'no2', 'rspm', 'spm', 'pm2_5']
for col in numerical_cols:
    df[col].fillna(df[col].mean(), inplace=True)
```

```python
print(df.isnull().sum())
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.dtypes
```

```python
df.drop(columns='sampling_date',axis=1,inplace=True)
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df.columns
```

```python
df['date'] = df['date'].str.replace('/', '-', regex=False)
```

```python
df['date'] = pd.to_datetime(df['date'], dayfirst=False)
```

```python
df['year'] = df['date'].dt.year
```

```python
df.columns
```

```python
df.head()
```

```python
df.sort_values(by='date', inplace=True)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['date'], df['so2'], marker='o', color='b', linestyle='-', linewidth=2, markersize=6)
#plt.plot(df['date'], df['no2'], marker='o', color='r', linestyle='-', linewidth=2, markersize=6)
```

```python
df.columns
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='year', y='so2', data=df)
plt.title('Average AQI by Year')
plt.xlabel('Year')
plt.ylabel('Average AQI')
plt.xticks(rotation=45)
plt.show()
```

```python
def categorize_pm2_5(pm25):
    if pm25 <= 35:
        return 'Low'
    elif pm25 <= 75:
        return 'Medium'
    else:
        return 'High'

df['pollutant_category'] = df['pm2_5'].apply(categorize_pm2_5)
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='pollutant_category', y='pm2_5', data=df)
plt.title('AQI Distribution by Pollutant Category')
plt.xlabel('Pollutant Category')
plt.ylabel('AQI')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x='year', y='so2', data=df)
plt.title('SO2 Distribution by Year')
plt.xlabel('Year')
plt.ylabel('SO2 Levels')
plt.xticks(rotation=40)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='year', y='so2', data=df)
plt.title('PM2.5 Distribution by Year')
plt.xlabel('Year')
plt.ylabel('PM2.5 Levels')
plt.xticks(rotation=40)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='pm2_5', y='so2', data=df)
plt.title('PM2.5 vs SO2 Levels')
plt.xlabel('PM2.5')
plt.ylabel('SO2')
plt.show()
```
